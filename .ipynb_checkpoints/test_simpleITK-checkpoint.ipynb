```python
import numpy as np
from skimage.io import imread, imsave
from skimage.data import cells3d
from skimage.filters import gaussian
from scipy.ndimage import convolve
from pyclesperanto_prototype import imshow
import pyclesperanto_prototype as cle
from napari_simpleitk_image_processing import richardson_lucy_deconvolution
import matplotlib.pyplot as plt
```

```python
from skimage import img_as_float
import imageio.v2 as iio
image = img_as_float(iio.imread("cells3d.tif"))[30,1,119:190,79:150]
# imshow(image)
```

```python
plt.subplot(1, 3, 1)
plt.imshow(image, cmap='gray')
plt.title('Original')
plt.axis('off')  # Ocultar los ejes para una mejor visualización
```

```python
perfect_psf = np.zeros( (71,71) )
perfect_psf[35, 35] = 1
psf = gaussian(perfect_psf, sigma=2)

plt.subplot(1, 3, 1)
plt.imshow(psf, cmap='gray')
plt.title('psf')
plt.axis('off')  # Ocultar los ejes para una mejor visualización
```

```python
convolved = convolve(image, psf)

plt.subplot(1, 3, 1)
plt.imshow(convolved, cmap='gray')
plt.title('Convolved')
plt.axis('off') 
```

```python
convolved.shape
```

```python
number_of_iterations = 30
deconvolved = richardson_lucy_deconvolution(convolved, psf, number_of_iterations)

plt.subplot(1, 3, 1)
plt.imshow(deconvolved, cmap='gray')
plt.title('Deconvolved')
plt.axis('off') 
```

```python
convolved.shape
```

```python
imsave("cells3d_convolved.tif",convolved)
imsave("cells3d_psf.tif",psf)
```

```python
%timeit -n 1 -r 10  richardson_lucy_deconvolution(convolved, psf, number_of_iterations)
```
