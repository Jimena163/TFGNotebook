```julia
using ImageCore, TestImages, ImageEdgeDetection, ImageFiltering
using JIVECore
```

```julia
path = "blobs.png"
im = JIVECore.Files.loadImage(path)
```

```julia
JIVECore.Process.imSmooth(im, 1)
```

```julia
JIVECore.Process.imSharpen(im, 1)
```

```julia
edges = JIVECore.Process.findEdges(im, true)
edges[1]
```

```julia
angles = JIVECore.Process.gradientOrientation(im)
mosaicview(im,Gray.(angles); nrow = 1)
```

```julia
thin = JIVECore.Process.computeThinnedEdges(Gray.(im))
mosaicview(im,edges[1],thin,edges[1].*thin; nrow = 1)
```

```julia
img_gray = Gray.(im)
```

```julia
using Plots
```

```julia
image = img_gray  # Assuming img_gray is already defined
local_maxima = JIVECore.Process.findPeaks(image; exclude_border=5, min_distance=5, threshold_rel=nothing, num_peaks=150, light_background=false)

# Display the image with local maxima
plot(image, c=:gray, label="Image")
scatter!([p[2] for p in local_maxima], [p[1] for p in local_maxima], color=:red, label="Local Maxima")
```
